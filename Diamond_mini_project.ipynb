## library and dataset

```r
install.packages("patchwork")
library(tidyverse)
glimpse(diamonds)
colnames(diamonds)
```

## 1. what is the relation of carat and price
the more carat increase,the more price increae no matter what clarity is

```r
ggplot(diamonds,aes(carat,price,col = price))+
  geom_point(alpha = 0.4)+
  geom_smooth(method = lm,col = "darkred")+
  theme_minimal()+
  scale_color_gradient(low = "gold",
                       high = "blue")+
  facet_wrap(~clarity,ncol = 2)
```

## 2. check overview of extream value on x,y,z by box plot
the minimum extream value on x,y,z have a value = 0 should check later what is it

```r
ggplot(diamonds,aes(x))+
  geom_boxplot()+
  #coord_flip()+
  theme_minimal()
ggplot(diamonds,aes(y))+
  geom_boxplot()+
  #coord_flip()+
  theme_minimal()
ggplot(diamonds,aes(z))+
  geom_boxplot()+
  #coord_flip()+
  theme_minimal()
```

## 3. What is frequency of the table variable in data set
the most table value are on 53-65

```r
ggplot(diamonds,aes(table,price,col = price))+
  geom_point(alpha = 0.5)+
  geom_rug()+
  scale_color_gradient(low = "blue",
                       high = "gold")+
  theme_minimal()

ggplot(diamonds,aes(table))+
  xlim(50,70)+
  ylim(0,20000)+
  geom_histogram(bins = 10)+
  theme_minimal()
```
